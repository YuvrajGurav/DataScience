# Scaling Implementation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/salary/Salary.csv')
df
```

**Normalization method 1- simple feature scaling**

data is rescaled and new values are [0,1]

Xnorm = (X-Xmin)/(Xmax-Xmin)

```python
df['Salary'].fillna((df['Salary'].mean()),inplace=True)
df['Age'].fillna((df['Age'].mean()),inplace=True)
df['Purchased']=df['Purchased'].apply(lambda x:0 if x=='No' else 1)
```

```python
df=pd.get_dummies(data=df, columns=['Country'])
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaled_features=scaler.fit_transform(df)
scaled_features
```

```python
df_MinMax = pd.DataFrame(data= scaled_features,columns=('Age','Salary','Purchased','Country_France','Country_Germany','Country_Spain'))
df_MinMax
```

**Standardization Z-score**

process of rescaling the features so that they will have properties of a gaussian distribution with mu=0 and sigma=1 where mu is the mean and sigma is the standard deviation

Z =(X-Xmin)/sigma: where sigma is standard deviation,/b>

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_x = sc_x.fit_transform(df)
sc_x
```

```python
sc_x = pd.DataFrame(data= sc_x,columns=('Age','Salary','Purchased','Country_France','Country_Germany','Country_Spain'))
sc_x
```
